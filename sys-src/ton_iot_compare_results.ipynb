# TON-IOT Model Comparison

This notebook:
1. Loads **saved inference data**
2. Compares **performance and efficiency** across of the inference data

```python
from pathlib import Path
from ton_iot_utils import load_and_compare_results, print_comparison_summary
```

```python
# Check if multiple result files exist
result_dir = Path('inference_results')
if result_dir.exists():
    result_files = list(result_dir.glob('*_inference_*.json'))
    
    if len(result_files) > 1:
        print(f"\n Found {len(result_files)} inference results. Comparing...")
        
        comparison_df = load_and_compare_results(result_files)
        print_comparison_summary(comparison_df)
        
        # Save comparison to CSV
        from datetime import datetime
        comparison_file = result_dir / f"model_comparison_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        comparison_df.to_csv(comparison_file, index=False)
        print(f"\n Comparison saved to: {comparison_file}")
        
    elif len(result_files) == 1:
        print(f"\nℹ Only one result file found. Run inference on more models to compare.")
    else:
        print(f"\nℹ No previous results found for comparison.")
else:
    print(f"\nℹ No inference_results directory found yet.")
```
